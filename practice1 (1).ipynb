# Creating a Simple Regression Model

```python
df.describe(include = 'all')
```

```python
df['date'] = pd.to_datetime(df['date'])
df.describe(include='all')
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
sns.lineplot(data = df, x = 'date', y = 'adj_close')
```

```python
df.columns
```

```python
# Filter numeric columns only
numeric_df = df.select_dtypes(include=[np.number])

# Compute correlation matrix on numeric data
correlation_matrix = numeric_df.corr()

# Display correlation matrix
correlation_matrix.style.background_gradient(cmap='coolwarm')#.format("{:.2f}")
```

```python
formula = 'adj_close ~ adj_close_1'
model = sm.ols(formula, data = df)
results = model.fit()
results.summary()
```

```python
df['pred_simple'] = results.predict(df)
```

```python
df[['date', 'adj_close', 'pred_simple']].head()
```

```python
dfs = df[df.date >= '2018-01-01']
dfs_long = dfs.melt(id_vars = 'date'
                    ,value_vars= ['adj_close', 'pred_simple']
                    , value_name = 'close'
                    , var_name='type')
dfs_long.sort_values('date').head()
```

```python
sns.lineplot(data = dfs_long, x='date',y='close',hue='type')
```

# <font color = 'red'>Multiple Regression Model</font>

```python
full_formula = 'adj_close~'
for i in range(len(df.columns)):
    if df.columns[i] in ['adj_close', 'date', 'ticker', 'pred_simple']:
        continue
    if full_formula[-1] != '~':
        full_formula += '+ '
    if df.columns[i] in ['dow']:
        full_formula += 'C(' + df.columns[i] + ')'
    else:
        full_formula += df.columns[i]
full_formula
```

```python
model_full = sm.ols(full_formula, df)
result_full = model_full.fit()
result_full.summary()
```

```python
df['pred_full'] = result_full.predict()
df[['date', 'adj_close', 'pred_simple', 'pred_full']].head()
```

```python
dfs = df[df.date >= '2018-01-01']
dfs_long = dfs.melt(id_vars = 'date'
                    ,value_vars= ['adj_close', 'pred_simple', 'pred_full']
                    , value_name = 'close'
                    , var_name='type')
dfs_long.sort_values('date').head()
```

```python
sns.lineplot(data = dfs_long, x = 'date', y = 'close', hue = 'type') 
```
